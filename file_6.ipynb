```python
import torch
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
```

Next, you will create an input tensor full of evenly spaced values on the interval [0,2ùúã]
 and specify requires_grad=True. (Like most functions that create tensors, torch.linspace() accepts an optional requires_grad option.) Setting this flag means that in every computation that follows, autograd will be accumulating the history of the computation in the output tensors of that computation

```python
a= torch.linspace(0., 2* math.pi, steps=25, requires_grad=True)
a
```

```python
#Next, you will perform a computation, and plot its output in terms of its inputs:
b = torch.sin(a)
print(b)
plt.plot(a.detach() , b.detach())
plt.xlabel("a")
plt.ylabel("b")
```
